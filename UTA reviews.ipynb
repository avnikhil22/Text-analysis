```python
#IMPORT COMMANDS
import pandas as pd
import nltk
```

```python
df = pd.read_excel("D:/UTA/Data Analysis Tools/Python/UTA Student Reviews.xlsx") # importing the file in jupyter
df.head()                   
```

```python
import string #import command
s = ""
for row in df.review[0: ]:     # I colleced all the reviews in a string 
    s+= row
(s)
```

```python
s1 = s.lower()               #all the lower cases
p =  string.punctuation      # commnd for removing puntuation
d = string.digits            # command for removing digits
p,d
```

```python
p_list = str.maketrans(p, len(p)*" ")    #code for putting space in the place of punctuation
d_list = str.maketrans(d, len(d)*" ")    #code for putting space in the place of digits 
```

```python
p_not = s1.translate(p_list) #punctuation free
pd_free = p_not.translate(d_list)#punctuation and digit free
#tokenize words
pd_free
```

```python
tok = nltk.word_tokenize(pd_free)   #tokenize command
#frequency of words
freq = nltk.FreqDist(tok)
(freq).plot(25)
```

```python
import string
s = ""
for row in df.review[0: ]:
    s+= row
s1 = s.lower()
p =  string.punctuation
d = string.digits
p_list = str.maketrans(p, len(p)*" ")
p_not = s1.translate(p_list) #punctuation free
d_list = str.maketrans(d, len(d)*" ")
pd_free = p_not.translate(d_list)#punctuation and digit free
#tokenize words
tok = nltk.word_tokenize(pd_free)
#frequency of words
freq = nltk.FreqDist(tok)
(freq)
```
